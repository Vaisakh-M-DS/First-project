```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data=pd.read_csv("employee.csv")
data
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
data
```

```python
from sklearn.preprocessing import LabelEncoder
dept = LabelEncoder()
data['department'] = dept.fit_transform(data['department'])

salary = LabelEncoder()
data['salary'] = salary.fit_transform(data['salary'])
```

```python
data.head()
```

```python
data.left.value_counts()
```

```python
x=data.drop('left',axis=1)
y=data['left']
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
smt = SMOTE()
x_train_smt, y_train_smt = smt.fit_resample(x_train, y_train)
```

```python
x_train_smt.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
# Instantiate the model
model1 = DecisionTreeClassifier()
model1.fit(x_train_smt, y_train_smt)
print('Train score:', model1.score(x_train_smt, y_train_smt))  
print('Test score:', model1.score(x_test, y_test))             
```

```python
 model1.get_depth()
```

```python
model2 = DecisionTreeClassifier(max_depth=21)
model2.fit(x_train_smt,y_train_smt)
print('Train score:',model2.score(x_train_smt,y_train_smt))
print('Test score:',model2.score(x_test,y_test))
```

```python
model2.get_depth()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
y_pred = model2.predict(x_test)
y_pred
```

```python
import pandas as pd
results_df = pd.DataFrame({"True Values": y_test, "Predicted Values": y_pred})
results_df
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

# Display the confusion matrix as a plot
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model2.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

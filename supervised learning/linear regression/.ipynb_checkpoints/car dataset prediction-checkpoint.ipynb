```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn. model_selection import train_test_split
```

```python
data = pd.read_csv("CAR DETAILS.CSV")
data
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
car_name=data["name"].unique()
fuel_type=data["fuel"].unique()
seller_type=data["seller_type"].unique()
transmission_type=data["transmission"].unique()
owner_type=data["owner"].unique()
```

```python
name = LabelEncoder()
data['name'] = name.fit_transform(data['name'])

fuel = LabelEncoder()
data['fuel'] = fuel.fit_transform(data['fuel'])

seller = LabelEncoder()
data['seller_type'] = seller.fit_transform(data['seller_type'])

transmission = LabelEncoder()
data['transmission'] = transmission.fit_transform(data['transmission'])

owner = LabelEncoder()
data['owner'] = owner.fit_transform(data['owner'])
```

```python
data
```

```python
x = data.drop('selling_price',axis = 1)
y = data['selling_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=90)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.score(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
pd.DataFrame({'True Value':y_test, 'Predicted Value': y_pred})
```

```python
plt.scatter(y_test,y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
```

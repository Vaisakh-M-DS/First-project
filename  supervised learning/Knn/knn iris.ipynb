```python
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("iris.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
vari = LabelEncoder()
data["variety"] = vari.fit_transform(data["variety"])
data
```

```python
data.variety.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x = data.drop("variety",axis = 1)
y= data['variety']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
smt = SMOTE()
x_train_smt, y_train_smt = smt.fit_resample(x_train_smt ,y_train_smt)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 5) # You can adjust the number of neighbors as needed
model.fit(x_train_smt, y_train_smt)
```

```python
model.score(x_train_smt, y_train_smt)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model1 = KNeighborsClassifier(n_neighbors = 5) 
model1.fit(x_train_smt, y_train_smt)
```

```python
model1.score(x_test,y_test)
```

```python
model.score(x_train_smt, y_train_smt)
```

```python
import matplotlib.pyplot as plt

# Plot the scores directly
plt.figure(figsize=(8, 5))
scores = [
    model.score(x_train_smt, y_train_smt),  # Training Accuracy
    model.score(x_test, y_test)                 # Testing Accuracy
]
labels = ['Training Accuracy', 'Testing Accuracy']

# Bar plot
plt.bar(labels, scores, color=['skyblue', 'orange'])
plt.ylim(0, 1.1)  # Set y-axis from 0 to 1.1 to account for annotation
plt.title("Model Performance: Training vs Testing Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Dataset")


# Show the plot
plt.show()
```
